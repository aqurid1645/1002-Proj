```python
import pandas as pd
```

```python
income = pd.read_csv("Cleansed/income.csv")

employment = pd.read_csv("Cleansed/employment.csv")

resignation = pd.read_csv("Cleansed/resignation.csv")

recruitment = pd.read_csv("Cleansed/recruitment.csv")

retrenchment = pd.read_csv("Cleansed/retrenchment.csv")

inflation = pd.read_csv("../Inflation/cpi_full.csv")

```

```python
inflation = inflation.filter(["Year","All Items "]).rename(columns={"Year":"year",
                                                                  "All Items ": "inflation"})
inflation
```

```python
combined = pd.merge(income, employment, on=["year", "industry1"], how ="outer")
combined = pd.merge(combined, resignation, on=["year", "industry1"], how ="outer")
combined = pd.merge(combined, recruitment, on=["year", "industry1"], how ="outer")
combined = pd.merge(combined, retrenchment, on=["year", "industry1"], how ="outer")
combined = pd.merge(combined, inflation, on="year", how ="outer")

combined = combined.rename(columns={"percent":"income_%_change",
                       "rate_x":"resignation_change",
                       "rate_y":"recruitment_change",
                       "retrench":"retrench_total"})
combined = combined.sort_values(by=["year"])
```

```python
combined
```

```python
combined.to_csv("Cleansed/combined.csv", index=False)
```
