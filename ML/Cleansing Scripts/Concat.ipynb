```python
import pandas as pd
from functools import reduce
```

```python
income = pd.read_csv("../Cleansed/income.csv")

employment = pd.read_csv("../Cleansed/employment.csv")

resignation = pd.read_csv("../Cleansed/resignation.csv")

recruitment = pd.read_csv("../Cleansed/recruitment.csv")

retrenchment = pd.read_csv("../Cleansed/retrenchment.csv")

inflation = pd.read_csv("../Raw Data/cpi_full.csv")

```

```python
inflation = inflation.filter(["Year","All Items "]).rename(columns={"Year":"year",
                                                                  "All Items ": "inflation"})
inflation
```

```python
concat_list = [income, employment, resignation, recruitment, retrenchment]
combined = reduce(lambda  left,right: pd.merge(left,right,on=["year", "industry1"]), concat_list)
combined = pd.merge(combined, inflation, on="year")

combined = combined.rename(columns={"percent":"income_%_change",
                       "rate_x":"resignation_change",
                       "rate_y":"recruitment_change",
                       "retrench":"retrench_total"})
combined = combined.sort_values(by=["year"]).dropna()
```

```python
combined
```

```python
combined.to_csv("../Cleansed/combined.csv", index=False)
```
