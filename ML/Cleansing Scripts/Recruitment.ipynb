```python
import pandas as pd
```

```python
recruitment = pd.read_csv("../Raw Data/Recruitment_rate.csv")
columns = recruitment.columns[2:].to_list()
recruitment
```

```python
values = []
for col in columns:
    print(col)
    print(recruitment[col].tolist())
    values.extend(recruitment[col].tolist())
len(values)
```

```python
new_col= []

for col in columns:
    new_col.extend([col] * 16)

years =[]
for i in range(2006, 2022):
    years.append(i)
years = years * 15

recruitment= pd.DataFrame({"year":years, "industry1":new_col, "rate":values})
recruitment
```
