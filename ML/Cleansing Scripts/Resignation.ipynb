```python
import pandas as pd
```

```python
resignation = pd.read_csv("../Raw Data/Recruitment_rate.csv")
columns = resignation.columns[2:].to_list()
```

```python
values = []
for col in columns:
    print(col)
    print(resignation[col].tolist())
    values.extend(resignation[col].tolist())
len(values)
```

```python
new_col= []

for col in columns:
    new_col.extend([col] * 16)

years =[]
for i in range(2006, 2022):
    years.append(i)
years = years * 15

resignation= pd.DataFrame({"year":years, "industry1":new_col, "rate":values})
resignation
```
